```python
# ============================================================
# PACKAGE 8.6.1 : RANDOM FOREST AQI FORECASTING MODEL
# ============================================================

import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score
)

import joblib


print("="*75)
print("PACKAGE 8.6.1 : RANDOM FOREST AQI FORECASTING MODEL")
print("="*75)



# ------------------------------------------------------------
# Load Selected Features
# ------------------------------------------------------------

X_train = pd.read_csv(
    r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Forecasting Model Pipeline\Package4_Feature_Selection\X_train_selected.csv"
)

X_test = pd.read_csv(
    r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Forecasting Model Pipeline\Package4_Feature_Selection\X_test_selected.csv"
)


y_train = pd.read_csv(
    r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Forecasting Model Pipeline\Package3_Feature_Encoding\y_train.csv"
).squeeze()


y_test = pd.read_csv(
    r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Forecasting Model Pipeline\Package3_Feature_Encoding\y_test.csv"
).squeeze()



print("\nDataset Shape")

print("X_train :", X_train.shape)

print("X_test  :", X_test.shape)

print("y_train :", y_train.shape)

print("y_test  :", y_test.shape)



# ------------------------------------------------------------
# Random Forest Model
# ------------------------------------------------------------

rf_model = RandomForestRegressor(

    n_estimators=300,

    max_depth=None,

    min_samples_split=2,

    min_samples_leaf=1,

    random_state=42,

    n_jobs=-1
)



print("\nTraining Random Forest Model...")


rf_model.fit(
    X_train,
    y_train
)


print("Training Completed")



# ------------------------------------------------------------
# Prediction
# ------------------------------------------------------------

y_pred = rf_model.predict(
    X_test
)



# ------------------------------------------------------------
# Evaluation Metrics
# ------------------------------------------------------------

rmse = np.sqrt(
    mean_squared_error(
        y_test,
        y_pred
    )
)


mae = mean_absolute_error(
    y_test,
    y_pred
)


r2 = r2_score(
    y_test,
    y_pred
)



results = pd.DataFrame(
    {
        "Model":
        [
            "Random Forest"
        ],

        "RMSE":
        [
            round(rmse,3)
        ],

        "MAE":
        [
            round(mae,3)
        ],

        "R2":
        [
            round(r2,4)
        ]
    }
)



print("\n============================================================")
print("RANDOM FOREST PERFORMANCE")
print("============================================================")

print(results)



# ------------------------------------------------------------
# Feature Importance
# ------------------------------------------------------------

feature_importance = pd.DataFrame(
    {
        "Feature":
        X_train.columns,

        "Importance":
        rf_model.feature_importances_
    }
)


feature_importance = (
    feature_importance
    .sort_values(
        by="Importance",
        ascending=False
    )
)



print("\n============================================================")
print("TOP 20 RANDOM FOREST FEATURES")
print("============================================================")


print(
    feature_importance.head(20)
)



# ------------------------------------------------------------
# Save Model
# ------------------------------------------------------------

joblib.dump(
    rf_model,
    "RandomForest_AQI_Model.pkl"
)



# ------------------------------------------------------------
# Save Predictions
# ------------------------------------------------------------

prediction_results = pd.DataFrame(
    {
        "Actual_AQI":
        y_test,

        "Predicted_AQI":
        y_pred
    }
)


prediction_results.to_csv(
    "RandomForest_Predictions.csv",
    index=False
)



# ------------------------------------------------------------
# Save Reports
# ------------------------------------------------------------

results.to_csv(
    "RandomForest_Performance.csv",
    index=False
)


feature_importance.to_csv(
    "RandomForest_Feature_Importance.csv",
    index=False
)



print("\nSaved Files:")
print("1. RandomForest_AQI_Model.pkl")
print("2. RandomForest_Predictions.csv")
print("3. RandomForest_Performance.csv")
print("4. RandomForest_Feature_Importance.csv")


print("\nPackage 8.6.1 Completed Successfully.")
```
