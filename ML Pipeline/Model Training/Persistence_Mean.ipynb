```python
# ============================================================
# PACKAGE 8.5 : BASELINE MODELS
# Persistence + Mean Baseline
# ============================================================

import pandas as pd
import numpy as np

from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score
)


print("="*75)
print("PACKAGE 8.5 : BASELINE MODELS")
print("="*75)



# ------------------------------------------------------------
# Load Selected Features
# ------------------------------------------------------------

X_train = pd.read_csv(
    r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Forecasting Model Pipeline\Package4_Feature_Selection\X_train_selected.csv"
)

X_test = pd.read_csv(
    r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Forecasting Model Pipeline\Package4_Feature_Selection\X_test_selected.csv"
)


y_train = pd.read_csv(
    r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Forecasting Model Pipeline\Package3_Feature_Encoding\y_train.csv"
).squeeze()


y_test = pd.read_csv(
    r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Forecasting Model Pipeline\Package3_Feature_Encoding\y_test.csv"
).squeeze()



print("\nDataset Shapes")

print(
    "X_train:",
    X_train.shape
)

print(
    "X_test:",
    X_test.shape
)

print(
    "y_train:",
    y_train.shape
)

print(
    "y_test:",
    y_test.shape
)



# ------------------------------------------------------------
# Evaluation Function
# ------------------------------------------------------------

def evaluate_model(
    name,
    y_true,
    y_pred
):

    rmse = np.sqrt(
        mean_squared_error(
            y_true,
            y_pred
        )
    )

    mae = mean_absolute_error(
        y_true,
        y_pred
    )

    r2 = r2_score(
        y_true,
        y_pred
    )


    return {
        "Model": name,
        "RMSE": round(rmse,3),
        "MAE": round(mae,3),
        "R2": round(r2,4)
    }



results = []



# ============================================================
# BASELINE 1 : PERSISTENCE MODEL
# ============================================================


print("\nRunning Persistence Baseline...")


if "AQI" not in X_test.columns:

    raise Exception(
        "AQI feature not found. Persistence baseline cannot run."
    )


persistence_prediction = X_test["AQI"]



persistence_result = evaluate_model(
    "Persistence (AQI_t = AQI_t-1)",
    y_test,
    persistence_prediction
)


results.append(
    persistence_result
)



# ============================================================
# BASELINE 2 : MEAN BASELINE
# ============================================================


print("Running Mean Baseline...")


mean_value = y_train.mean()


mean_prediction = np.full(
    shape=len(y_test),
    fill_value=mean_value
)



mean_result = evaluate_model(
    "Mean Baseline",
    y_test,
    mean_prediction
)


results.append(
    mean_result
)



# ------------------------------------------------------------
# Results
# ------------------------------------------------------------

results_df = pd.DataFrame(
    results
)



print("\n============================================================")
print("BASELINE PERFORMANCE")
print("============================================================")


print(
    results_df
)



# ------------------------------------------------------------
# Save Results
# ------------------------------------------------------------

results_df.to_csv(
    "Baseline_Model_Results.csv",
    index=False
)



print("\nSaved File:")
print("Baseline_Model_Results.csv")


print("\nPackage 8.5 Completed Successfully.")
```
