```python
# ============================================================
# PACKAGE 12 : CITIZEN HEALTH ADVISORY AGENT
# VERSION 3.0
# PART 3/3
#
# Purpose:
# - Generate visualization outputs
# - Create metadata report
# - Finalize Package 12
#
# Input:
#   Citizen_Health_Advisory_Report.csv
#
# Output:
#   Health_Risk_Distribution.png
#   Package12_Metadata.txt
#   Package12_Final_Summary.txt
#
# ============================================================


import os
import pandas as pd
import matplotlib.pyplot as plt



print("="*75)
print("PACKAGE 12 VERSION 3.0")
print("CITIZEN HEALTH ADVISORY AGENT")
print("PART 3/3")
print("="*75)



# ============================================================
# CONFIGURATION
# ============================================================


OUTPUT_DIR = (
    r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Citizen_Health_Advisory\Part3"
)


REPORT_FILE = (r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Citizen_Health_Advisory\Part2\Citizen_Health_Advisory_Report.csv")

RISK_PLOT = os.path.join(
    OUTPUT_DIR,
    "Health_Risk_Distribution.png"
)

METADATA_FILE = os.path.join(
    OUTPUT_DIR,
    "Package12_Metadata.txt"
)

SUMMARY_FILE = os.path.join(
    OUTPUT_DIR,
    "Package12_Final_Summary.txt"
)



# ============================================================
# LOAD FINAL REPORT
# ============================================================


print("\nLoading Citizen Advisory Report...")


if not os.path.exists(REPORT_FILE):

    raise FileNotFoundError(
        "Citizen_Health_Advisory_Report.csv not found. Run Part 2 first."
    )



health_df = pd.read_csv(
    REPORT_FILE
)



print(
    "Dataset Shape:",
    health_df.shape
)



# ============================================================
# HEALTH RISK SUMMARY
# ============================================================


print("\nHealth Risk Distribution")

risk_distribution = (

    health_df["Health_Risk_Level"]

    .value_counts()

)



print(
    risk_distribution
)



# ============================================================
# VISUALIZATION
# ============================================================


plt.figure(
    figsize=(8,5)
)


risk_distribution.plot(
    kind="bar"
)


plt.xlabel(
    "Health Risk Level"
)


plt.ylabel(
    "Number of Wards"
)


plt.title(
    "Delhi Ward Health Risk Distribution"
)


plt.tight_layout()


plt.savefig(
    RISK_PLOT,
    dpi=300
)


plt.close()



print(
    "\nSaved:"
)

print(
    RISK_PLOT
)



# ============================================================
# TOP HIGH RISK WARDS
# ============================================================


top_risk = health_df.sort_values(

    by="Health_Risk_Score",

    ascending=False

).head(10)



# ============================================================
# METADATA GENERATION
# ============================================================



with open(
    METADATA_FILE,
    "w",
    encoding="utf-8"
) as f:


    f.write(
        "=================================================\n"
    )

    f.write(
        "PACKAGE 12 : CITIZEN HEALTH ADVISORY AGENT\n"
    )

    f.write(
        "=================================================\n\n"
    )


    f.write(
        f"Total Wards Analysed : {len(health_df)}\n"
    )


    f.write(
        f"Average Health Risk Score : "
        f"{health_df['Health_Risk_Score'].mean():.2f}\n\n"
    )


    f.write(
        "Risk Distribution\n"
    )

    f.write(
        "-----------------\n"
    )


    for k,v in risk_distribution.items():

        f.write(
            f"{k} : {v} wards\n"
        )


    f.write(
        "\n\nTop 10 High Risk Wards\n"
    )

    f.write(
        "----------------------\n"
    )


    for _,row in top_risk.iterrows():


        f.write(

            f"{row.get('Ward_Name','Unknown')} | "

            f"Risk Score: "
            f"{row['Health_Risk_Score']:.2f} | "

            f"Source: "
            f"{row['Dominant_Source']}\n"

        )



print(
    "Saved:"
)

print(
    METADATA_FILE
)



# ============================================================
# FINAL SUMMARY FILE
# ============================================================



critical_count = (

    health_df[

        health_df["Health_Risk_Level"]

        ==
        "Critical"

    ]

    .shape[0]

)



high_count = (

    health_df[

        health_df["Health_Risk_Level"]

        ==
        "High"

    ]

    .shape[0]

)



with open(

    SUMMARY_FILE,

    "w",

    encoding="utf-8"

) as f:


    f.write(
        "=================================================\n"
    )

    f.write(
        "PACKAGE 12 COMPLETION SUMMARY\n"
    )

    f.write(
        "CITIZEN HEALTH ADVISORY AGENT\n"
    )

    f.write(
        "=================================================\n\n"
    )


    f.write(
        "Objective:\n"
    )

    f.write(
        "Generate ward-level health risk advisories "
        "using AQI forecast, pollution source attribution "
        "and sensitive location vulnerability.\n\n"
    )


    f.write(
        f"Wards Analysed: {len(health_df)}\n"
    )


    f.write(
        f"Critical Risk Wards: {critical_count}\n"
    )


    f.write(
        f"High Risk Wards: {high_count}\n\n"
    )


    f.write(
        "Generated Outputs:\n"
    )

    f.write(
        "- Ward_Health_Risk.csv\n"
    )

    f.write(
        "- Citizen_Health_Advisory_Report.csv\n"
    )

    f.write(
        "- Frontend_Health_Advisory.json\n"
    )

    f.write(
        "- Health_Risk_Distribution.png\n"
    )

    f.write(
        "- Package12_Metadata.txt\n"
    )



print(
    "Saved:"
)

print(
    SUMMARY_FILE
)



print("\n")
print("="*75)
print("PACKAGE 12 COMPLETED SUCCESSFULLY")
print("="*75)



print(
    """
Generated Files:

1. Ward_Health_Risk.csv

2. Citizen_Health_Advisory_Report.csv

3. Frontend_Health_Advisory.json

4. Health_Risk_Distribution.png

5. Package12_Metadata.txt

6. Package12_Final_Summary.txt

"""
)
```
