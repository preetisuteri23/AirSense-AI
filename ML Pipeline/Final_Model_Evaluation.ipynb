```python
# ============================================================
# PACKAGE 9.7A : FINAL MODEL EVALUATION
# ============================================================


import pandas as pd
import numpy as np

import joblib
import os


from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score,
    accuracy_score,
    classification_report
)



print("="*75)
print("PACKAGE 9.7A : FINAL MODEL EVALUATION")
print("="*75)



# ------------------------------------------------------------
# Paths
# ------------------------------------------------------------

BASE_PATH = (
r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Forecasting Model Pipeline"
)



MODEL_PATH = (
r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Last_Stage\Final_XGBoost_Production_Training\Final_XGBoost_AQI_Model.pkl"
)



X_TEST_PATH = (
BASE_PATH +
r"\Package4_Feature_Selection\X_test_selected.csv"
)



Y_TEST_PATH = (
BASE_PATH +
r"\Package3_Feature_Encoding\y_test.csv"
)



OUTPUT_FOLDER = (
r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Last_Stage\Final_Model_Evaluation"
)



os.makedirs(
    OUTPUT_FOLDER,
    exist_ok=True
)



# ------------------------------------------------------------
# Load Model
# ------------------------------------------------------------


print("\nLoading Production Model...")


model = joblib.load(
    MODEL_PATH
)


print("Model Loaded Successfully")



# ------------------------------------------------------------
# Load Test Data
# ------------------------------------------------------------


X_test = pd.read_csv(
    X_TEST_PATH
)



y_test = pd.read_csv(
    Y_TEST_PATH
).squeeze()



print("\nDataset Shape")

print(
"X_test:",
X_test.shape
)

print(
"y_test:",
y_test.shape
)



# ------------------------------------------------------------
# Prediction
# ------------------------------------------------------------


print("\nGenerating Predictions...")


y_pred = model.predict(
    X_test
)



# ------------------------------------------------------------
# Regression Metrics
# ------------------------------------------------------------


rmse = np.sqrt(
    mean_squared_error(
        y_test,
        y_pred
    )
)



mae = mean_absolute_error(
    y_test,
    y_pred
)



r2 = r2_score(
    y_test,
    y_pred
)



metrics = pd.DataFrame(

{

"Metric":

[
"RMSE",
"MAE",
"R2"
],


"Value":

[
rmse,
mae,
r2
]

}

)



metrics.to_csv(

OUTPUT_FOLDER+
r"\Final_Evaluation_Metrics.csv",

index=False

)



# ------------------------------------------------------------
# Prediction Output
# ------------------------------------------------------------


prediction_df = pd.DataFrame(

{

"Actual_AQI":
y_test,


"Predicted_AQI":
y_pred,


"Error":
y_test-y_pred,


"Absolute_Error":
abs(y_test-y_pred)

}

)



prediction_df.to_csv(

OUTPUT_FOLDER+
r"\Final_Prediction_Results.csv",

index=False

)



# ------------------------------------------------------------
# High Error Cases
# ------------------------------------------------------------


high_error = (

prediction_df

.sort_values(

by="Absolute_Error",

ascending=False

)

.head(20)

)



high_error.to_csv(

OUTPUT_FOLDER+
r"\Final_High_Error_Cases.csv",

index=False

)



# ------------------------------------------------------------
# AQI Category Evaluation
# ------------------------------------------------------------


def aqi_category(aqi):

    if aqi <= 50:
        return "Good"

    elif aqi <=100:
        return "Satisfactory"

    elif aqi <=200:
        return "Moderate"

    elif aqi <=300:
        return "Poor"

    elif aqi <=400:
        return "Very Poor"

    else:
        return "Severe"



category_df = pd.DataFrame(

{

"Actual_AQI":
y_test,


"Predicted_AQI":
y_pred

}

)



category_df["Actual_Category"] = (

category_df["Actual_AQI"]
.apply(aqi_category)

)



category_df["Predicted_Category"] = (

category_df["Predicted_AQI"]
.apply(aqi_category)

)



category_accuracy = accuracy_score(

category_df["Actual_Category"],

category_df["Predicted_Category"]

)



category_df.to_csv(

OUTPUT_FOLDER+
r"\Final_AQI_Category_Evaluation.csv",

index=False

)



print("\n============================================================")

print("FINAL PERFORMANCE")

print("============================================================")


print(
metrics
)



print(
"\nAQI Category Accuracy:",
round(category_accuracy,4)
)



print("\nPACKAGE 9.7A COMPLETED SUCCESSFULLY")
```
