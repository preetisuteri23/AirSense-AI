```python
# ============================================================
# PACKAGE 8.1 : CHRONOLOGICAL TRAIN-TEST SPLIT
# ============================================================

import pandas as pd


print("="*60)
print("PACKAGE 8.1 : CHRONOLOGICAL TRAIN-TEST SPLIT")
print("="*60)


# ------------------------------------------------------------
# Load Forecasting Dataset
# ------------------------------------------------------------

file_path = r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Forecast_Dataset_Leakage_Removal\Delhi_Forecasting_Dataset.csv"

df = pd.read_csv(file_path)


# ------------------------------------------------------------
# Date Conversion
# ------------------------------------------------------------

df["Date"] = pd.to_datetime(df["Date"])

# Sort chronologically
df = df.sort_values("Date").reset_index(drop=True)


print("\nOriginal Dataset Shape:")
print(df.shape)


print("\nDate Range:")
print(df["Date"].min())
print("to")
print(df["Date"].max())


# ------------------------------------------------------------
# Define Split Date
# ------------------------------------------------------------

split_date = "2020-01-01"


# ------------------------------------------------------------
# Chronological Split
# ------------------------------------------------------------

train = df[df["Date"] < split_date]

test = df[df["Date"] >= split_date]


# ------------------------------------------------------------
# Separate Features and Target
# ------------------------------------------------------------

TARGET = "AQI_Target"


X_train = train.drop(columns=[TARGET])

y_train = train[TARGET]


X_test = test.drop(columns=[TARGET])

y_test = test[TARGET]


# ------------------------------------------------------------
# Results
# ------------------------------------------------------------

print("\n============================================================")
print("TRAIN TEST SPLIT RESULT")
print("============================================================")


print("\nTraining Data Shape:")
print(X_train.shape)

print("Training Target Shape:")
print(y_train.shape)


print("\nTesting Data Shape:")
print(X_test.shape)

print("Testing Target Shape:")
print(y_test.shape)


print("\nTraining Date Range:")
print(train["Date"].min())
print("to")
print(train["Date"].max())


print("\nTesting Date Range:")
print(test["Date"].min())
print("to")
print(test["Date"].max())


# ------------------------------------------------------------
# Save Split Data
# ------------------------------------------------------------

train.to_csv(
    "Delhi_AQI_Train.csv",
    index=False
)

test.to_csv(
    "Delhi_AQI_Test.csv",
    index=False
)


print("\nSaved Files:")
print("1. Delhi_AQI_Train.csv")
print("2. Delhi_AQI_Test.csv")


print("\nPackage 8.1 Completed Successfully.")
```
