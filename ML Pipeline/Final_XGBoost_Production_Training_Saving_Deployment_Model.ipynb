```python
# ============================================================
# PACKAGE 9.6 : FINAL XGBOOST PRODUCTION TRAINING
# ============================================================


import pandas as pd
import numpy as np

from xgboost import XGBRegressor

import joblib
import json
import os


print("="*75)
print("PACKAGE 9.6 : FINAL XGBOOST PRODUCTION TRAINING")
print("="*75)



# ------------------------------------------------------------
# Paths
# ------------------------------------------------------------

BASE_PATH = r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Forecasting Model Pipeline"


X_TRAIN_PATH = (
    BASE_PATH +
    r"\Package4_Feature_Selection\X_train_selected.csv"
)


X_TEST_PATH = (
    BASE_PATH +
    r"\Package4_Feature_Selection\X_test_selected.csv"
)


Y_TRAIN_PATH = (
    BASE_PATH +
    r"\Package3_Feature_Encoding\y_train.csv"
)


Y_TEST_PATH = (
    BASE_PATH +
    r"\Package3_Feature_Encoding\y_test.csv"
)



OUTPUT_FOLDER = (
   r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Last_Stage\Final_XGBoost_Production_Training"
)


os.makedirs(
    OUTPUT_FOLDER,
    exist_ok=True
)



# ------------------------------------------------------------
# Load Dataset
# ------------------------------------------------------------

print("\nLoading Selected Features...")


X_train = pd.read_csv(
    X_TRAIN_PATH
)


X_test = pd.read_csv(
    X_TEST_PATH
)



y_train = pd.read_csv(
    Y_TRAIN_PATH
).squeeze()



y_test = pd.read_csv(
    Y_TEST_PATH
).squeeze()



print("\nDataset Shapes")

print("X_train :", X_train.shape)

print("X_test  :", X_test.shape)

print("y_train :", y_train.shape)

print("y_test  :", y_test.shape)



# ------------------------------------------------------------
# Combine Full Dataset
# ------------------------------------------------------------

print("\nCombining Training + Testing Data...")


X_full = pd.concat(
    [
        X_train,
        X_test
    ],
    axis=0
).reset_index(
    drop=True
)



y_full = pd.concat(
    [
        y_train,
        y_test
    ],
    axis=0
).reset_index(
    drop=True
)



print("\nFinal Production Dataset")

print(
    "X_full :",
    X_full.shape
)

print(
    "y_full :",
    y_full.shape
)



# ------------------------------------------------------------
# Final XGBoost Model
# ------------------------------------------------------------


print("\nInitializing Final XGBoost Model...")


final_xgb_model = XGBRegressor(

    n_estimators=500,

    learning_rate=0.03,

    max_depth=4,

    subsample=0.8,

    colsample_bytree=0.8,

    reg_lambda=5,

    objective="reg:squarederror",

    random_state=42,

    n_jobs=-1
)



# ------------------------------------------------------------
# Training
# ------------------------------------------------------------


print("\nTraining Production Model...")


final_xgb_model.fit(
    X_full,
    y_full
)


print("Production Training Completed")



# ------------------------------------------------------------
# Save Deployment Model
# ------------------------------------------------------------


MODEL_PATH = (
    OUTPUT_FOLDER +
    r"\Final_XGBoost_AQI_Model.pkl"
)



joblib.dump(
    final_xgb_model,
    MODEL_PATH
)



# ------------------------------------------------------------
# Save Feature List
# ------------------------------------------------------------


feature_file = (
    OUTPUT_FOLDER +
    r"\Final_Model_Features.csv"
)



features = pd.DataFrame(
    {
        "Feature":
        X_full.columns
    }
)



features.to_csv(
    feature_file,
    index=False
)



# ------------------------------------------------------------
# Save Model Parameters
# ------------------------------------------------------------


parameter_file = (
    OUTPUT_FOLDER +
    r"\Final_Model_Parameters.json"
)



parameters = {

    "model":
    "XGBoost AQI Forecasting Model",

    "n_estimators":
    500,

    "learning_rate":
    0.03,

    "max_depth":
    4,

    "subsample":
    0.8,

    "colsample_bytree":
    0.8,

    "reg_lambda":
    5,

    "objective":
    "reg:squarederror",

    "random_state":
    42,

    "training_samples":
    int(X_full.shape[0]),

    "feature_count":
    int(X_full.shape[1])
}



with open(
    parameter_file,
    "w"
) as file:

    json.dump(
        parameters,
        file,
        indent=4
    )



# ------------------------------------------------------------
# Training Report
# ------------------------------------------------------------


report_file = (
    OUTPUT_FOLDER +
    r"\Final_Model_Training_Report.csv"
)



report = pd.DataFrame(
    {

        "Parameter":

        [
            "Training Samples",
            "Features",
            "Model",
            "Validation RMSE",
            "Validation MAE",
            "Validation R2"
        ],


        "Value":

        [
            X_full.shape[0],
            X_full.shape[1],
            "XGBoost",
            28.1797,
            21.5862,
            0.8966
        ]

    }
)



report.to_csv(
    report_file,
    index=False
)



# ------------------------------------------------------------
# Final Output
# ------------------------------------------------------------


print("\n============================================================")
print("FINAL PRODUCTION MODEL SAVED")
print("============================================================")


print(
    "1.",
    MODEL_PATH
)

print(
    "2.",
    feature_file
)

print(
    "3.",
    parameter_file
)

print(
    "4.",
    report_file
)



print("\nPACKAGE 9.6 COMPLETED SUCCESSFULLY")
```
