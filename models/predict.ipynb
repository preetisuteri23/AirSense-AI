```python
# ============================================================
# predict.py
#
# Production Inference API
# AI-Powered Urban Air Quality Intelligence Platform
#
# Model:
# XGBoost AQI Forecasting Model
#
# Input:
# Dictionary / JSON containing selected ML features
#
# Output:
# Predicted AQI + AQI Category
#
# ============================================================


import os
import pickle
import json
import pandas as pd



# ============================================================
# PATH CONFIGURATION
# ============================================================


MODEL_FILE = (r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Last_Stage\Final_XGBoost_Production_Training\Final_XGBoost_AQI_Model.pkl")


FEATURE_FILE = (r"C:\Users\91863\Desktop\AIR QUALITY INTELLIGENCE\Forecasting Model Pipeline\Package4_Feature_Selection\Selected_Features.pkl")


# ============================================================
# LOAD MODEL AND FEATURES
# ============================================================


def load_model():


    with open(
        MODEL_FILE,
        "rb"
    ) as f:

        model = pickle.load(f)



    with open(
        FEATURE_FILE,
        "rb"
    ) as f:

        selected_features = pickle.load(f)



    return model, selected_features





model, selected_features = load_model()



print("="*60)
print("XGBoost AQI Production Model Loaded")
print("="*60)

print(
    "Number of Features:",
    len(selected_features)
)





# ============================================================
# AQI CLASSIFICATION
# ============================================================


def get_aqi_category(aqi):


    if aqi <= 50:

        return "Good"


    elif aqi <= 100:

        return "Satisfactory"


    elif aqi <= 200:

        return "Moderate"


    elif aqi <= 300:

        return "Poor"


    elif aqi <= 400:

        return "Very Poor"


    else:

        return "Severe"





# ============================================================
# MAIN PREDICTION FUNCTION
# ============================================================


def predict_aqi(input_data):


    """
    input_data:

    Dictionary containing model features

    Example:

    {
        "PM2.5": 120,
        "PM10": 200,
        "NO2": 40,
        ...
    }

    """



    # Convert input dictionary to dataframe

    input_df = pd.DataFrame(
        [input_data]
    )



    # --------------------------------------------------------
    # Validate Features
    # --------------------------------------------------------


    missing_features = [

        feature

        for feature in selected_features

        if feature not in input_df.columns

    ]



    if missing_features:


        raise ValueError(

            "Missing required features: "

            +
            str(missing_features)

        )



    # --------------------------------------------------------
    # Feature Ordering
    # VERY IMPORTANT
    # --------------------------------------------------------


    input_df = input_df[
        selected_features
    ]



    # --------------------------------------------------------
    # Prediction
    # --------------------------------------------------------


    prediction = model.predict(
        input_df
    )[0]



    prediction = float(
        prediction
    )



    # --------------------------------------------------------
    # Response
    # --------------------------------------------------------


    result = {


        "model":

        "XGBoost AQI Forecasting Model",



        "feature_count":

        len(selected_features),



        "predicted_AQI":

        round(
            prediction,
            2
        ),



        "AQI_Category":

        get_aqi_category(
            prediction
        )

    }



    return result





# ============================================================
# LOCAL TEST
# ============================================================


if __name__ == "__main__":


    print("\nTesting Model...\n")



    # Dummy test
    # Replace values from frontend


    sample_input = {}



    try:


        output = predict_aqi(
            sample_input
        )


        print(
            json.dumps(
                output,
                indent=4
            )
        )


    except Exception as e:


        print(
            "Prediction Test Skipped:"
        )

        print(e)
```
